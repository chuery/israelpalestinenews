```python
# read in data

import pandas as pd

df_gaza = pd.read_csv("gaza.csv")
df_gaza_two = pd.read_csv("gaza_two.csv")
df_israel = pd.read_csv("israel.csv")
df_israel_palestine = pd.read_csv("israel_palestine.csv")
df_israel_two = pd.read_csv("israel_two.csv")
df_palestine = pd.read_csv("palestine.csv")
df_gaza_three = pd.read_csv("gaza_three-2.csv")
df_palestinians = pd.read_csv("palestinians.csv")
df_idf = pd.read_csv("idf.csv")
df_hamas = pd.read_csv("hamas.csv")
df_hamas_two = pd.read_csv("hamas_two.csv")
df_hamas_three = pd.read_csv("hamas_three.csv")
```

```python
# combine searches into one dataframe, remove duplicates
df_headlines = pd.concat([df_gaza,
                          df_gaza_two,
                          df_israel,
                          df_israel_palestine,
                          df_israel_two,
                          df_palestine,
                          df_gaza_three,
                          df_palestinians,
                          df_idf,
                          df_hamas,
                          df_hamas_two,
                          df_hamas_three,
                          ]
                         ).drop_duplicates().reset_index(drop=True)
df_headlines
```

```python
# format date, sort values
df_headlines['Date'] = pd.to_datetime(df_headlines['Date'])
df_headlines_sorted =  df_headlines.sort_values(by='Date')
df_headlines_sorted = df_headlines_sorted[["Date", "Headline"]].drop_duplicates(subset=['Date', 'Headline']).reset_index(drop=True)
df_headlines_sorted
```

```python
# save sorted combined dataframe
from google.colab import files
df_headlines_sorted.to_csv("headlines_sorted.csv")
files.download('headlines_sorted.csv')
```
