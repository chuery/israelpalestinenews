```python
# read in data
import pandas as pd

df_gaza = pd.read_csv("guardian_gaza.csv")
df_israel = pd.read_csv("guardian_israel-2.csv")
df_israel_palestine = pd.read_csv("guardian_israel_palestine.csv")
df_palestine = pd.read_csv("guardian_palestine.csv")
df_palestinians = pd.read_csv("guardian_palestinians.csv")
df_idf = pd.read_csv("guardian_idf.csv")
df_hamas = pd.read_csv("guardian_hamas.csv")
```

```python
# combine searches into one dataframe, remove duplicates
df_headlines = pd.concat([df_gaza,
                          df_israel,
                          df_israel_palestine,
                          df_palestine,
                          df_palestinians,
                          df_idf,
                          df_hamas,
                          ]
                         ).drop_duplicates().reset_index(drop=True)

df_headlines
```

```python
# reformat dataframes to ensure correct ordering
if 'Date' in df_headlines.columns and 'Headline' in df_headlines.columns:
    df_headlines = df_headlines.rename(columns={'Date': 'hold', 'Headline': 'Date'})
    df_headlines = df_headlines.rename(columns={'hold': 'Headline'})

df_headlines = df_headlines[["Date", "Headline"]]
df_headlines
```

```python
# convert correct date format, ensure duplicates are removed
df_headlines['Date'] = pd.to_datetime(df_headlines['Date'])
df_headlines_sorted =  df_headlines.sort_values(by='Date')
df_headlines_sorted = df_headlines_sorted[["Date", "Headline"]].drop_duplicates(subset=['Date', 'Headline']).reset_index(drop=True)
df_headlines_sorted
```

```python
# download combined/sorted headlines
from google.colab import files
df_headlines_sorted.to_csv("guardian_headlines_sorted.csv")
files.download('guardian_headlines_sorted.csv')
```
