```python
# search: "Israel Palestine", 100 pages
import pandas as pd
import requests
import json
import time

LILY_API_KEY = "jJi7vlOqXlXStL77MGOGD1kOT3cGXzuV"
BASE_URL = "https://api.nytimes.com/svc/search/v2/articlesearch.json"
QUERY = "Israel Palestine"
START_DATE = "20231001"  # YYYYMMDD format
END_DATE = "20250228"
PAGE_LIMIT = 100  # Max limit (100 pages allowed, adjust as needed)
REQUEST_DELAY = 12  # Seconds between requests to avoid spamming

def fetch_articles():
    articles_data = []
    for page in range(PAGE_LIMIT):
        params = {
            "q": QUERY,
            "begin_date": START_DATE,
            "end_date": END_DATE,
            "api-key": LILY_API_KEY,
            "page": page
        }
        try:
            response = requests.get(BASE_URL, params=params)
            response.raise_for_status()
            data = response.json()
            docs = data.get("response", {}).get("docs", [])
            for article in docs:
                headline = article.get("headline", {}).get("main", "No Title")
                pub_date = article.get("pub_date", "No Date")
                articles_data.append((pub_date, headline))
        except requests.exceptions.RequestException as e:
            print(f"Request error: {e}")
            break

        time.sleep(REQUEST_DELAY)  # Prevents exceeding rate limits

    return articles_data

if __name__ == "__main__":
    articles = fetch_articles()
    df_israel_palestine = pd.DataFrame(articles, columns=["Date", "Headline"])  # Create DataFrame

df_israel_palestine
```
